```python
def convertir(df):

    
    # Itera sobre las columnas y si son de tipo 'object' (strings), las convierte a minúsculas y reemplaza espacios por '_'
    for columna in df.columns:
        if df[columna].dtype == 'object':  # Verifica si la columna es de tipo 'object' (cadena)
            df[columna] = df[columna].str.lower().str.replace(' ', '_')
    
    return df
```

```python
import re
```

```python
def convertir_a_numerico(valor):
    # Utilizamos expresiones regulares para extraer los números y las unidades (meses o años)
    num = re.findall(r'\d+', valor)[0]  # Encuentra el número en el string
    unidad = valor.split(' ')[-1]  # Encuentra la unidad (months o years)

    if unidad == 'months' or unidad == 'month':
        return int(num)
    elif unidad == 'year' or unidad == 'years':
        return int(num) * 12  # Convertir años a meses

# Aplicar la función a la columna 'monthly_duration_of_opioid_medication'
df["monthly_duration_of_opioid_medication"] = df["monthly_duration_of_opioid_medication"].apply(convertir_a_numerico)

print(df)
```

```python
def extract_dosage_frequency(text):
    dosage_pattern = r'(\d+)(mg|mcg)'
    frequency_pattern = r'(\d+)\s*times per day'

    dosage_matches = re.findall(dosage_pattern, text)
    frequency_matches = re.findall(frequency_pattern, text)

    dosages = [' '.join(match) for match in dosage_matches]
    frequency = frequency_matches[0] if frequency_matches else None

    return dosages, frequency

# Aplicar la función a la columna 'dosage_and_frequency_of_opioid_medication_use'
df[['dosage_opioid_medication', 'frequency_medication_per_day']] = df['dosage_and_frequency_of_opioid_medication_use'].apply(extract_dosage_frequency).apply(pd.Series)

print(df[['dosage_and_frequency_of_opioid_medication_use', 'dosage_opioid_medication', 'frequency_medication_per_day']])
```

```python
def cambiar_none_a_valor(valor):
    if valor is None:
        return 0.33
    else:
        return valor

# Aplicar la función a la columna 'Frequency' del DataFrame
df['frequency_medication_per_day'] = df['frequency_medication_per_day'].apply(cambiar_none_a_valor)
```
